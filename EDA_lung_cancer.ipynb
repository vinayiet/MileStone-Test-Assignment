```python
df = df.drop_duplicates()
df
```

```python
import matplotlib.pyplot as plt
df.hist(bins = 50, figsize=(20,15))
plt.show()
```

```python
# Univariate analysis: In this analysis we use single feature to do analysis
import seaborn as sns
plt.figure(figsize= (15,10))
sns.boxplot(data=df.select_dtypes(include=[int, float]))
plt.xticks(rotation=45)
plt.show
```

```python
# Analysis using categorical columns 
categorical_columns = df.select_dtypes(include = [object]).columns
for categorical_column in categorical_columns:
    plt.figure(figsize = (8,5))
    sns.countplot(x=categorical_column, data=df)
    plt.title(categorical_column)
    plt.show()
```

```python
# Bivariate analysis : In this analysis we take two feature to do analysis
correlation_matrix = df.corr()
plt.figure(figsize = (10,15))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
```

```python
sns.pairplot(df)
plt.show()
```

```python
grouped_data = df.groupby('LUNG_CANCER').mean()
print(grouped_data)
```

```python
# Visualizations for multivariate analysis
plt.figure(figsize=(15, 10))
sns.pairplot(df, hue='LUNG_CANCER')
plt.show()
```

```python
df['LUNG_CANCER'].value_counts()
```

```python
for column in df.columns:
    if column != 'LUNG_CANCER' and df[column].dtype in ['int64', 'float64']:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='LUNG_CANCER', y=column, data=df)
        plt.title(f'{column} vs LUNG_CANCER')
        plt.show()
```

```python
from sklearn.preprocessing import StandardScaler# Feature Engineering
df['NEW_FEATURE'] = df['AGE'] * df['SMOKING']

scaler = StandardScaler()
df[['AGE', 'SMOKING', 'YELLOW_FINGERS']] = scaler.fit_transform(df[['AGE', 'SMOKING', 'YELLOW_FINGERS']])

# Save the cleaned and transformed data
df.to_csv('cleaned_lung_cancer_data.csv', index=False)
```
